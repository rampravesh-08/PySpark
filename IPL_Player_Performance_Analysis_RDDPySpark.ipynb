```python
def remove_first_rec(line):
    fields = line.split(",")
    if fields[0] == "Team" and fields[1] == "Player":
        pass
    else:
        return line

#ipl_data_rdd = ipl_data.filter(lambda line: remove_first_rec(line))
ipl_data_rdd = ipl_data.filter(lambda line: not line.startswith('Team'))
```

```python
ipl_data_rdd.take(2)
```

```python
ipl_data_flatten = ipl_data_rdd.flatMap(lambda line: line.split(","))
ipl_data_flatten.collect()
```

```python
fields = ("Team","Player","Tournament","Matches","Batting_Innings","Not_Out","Runds_Scored","Highest_Score","Batting_Average","Balls_Faced","Batting_Strike_Rate","Century","Half_Century","Zero","Fours","Sixes","Bowling_Innings","Overs_Bowled","Maidens_Bowled","Runs_Conceded","Wickets_Taken","Best_Bowling_Figures","Bowling_Average","Bowling_Economy_Rate","Bowling_Strike_Rate","FourPlus_Innings_Wickets","FivePlus_Innings_Wickets","Catches_Taken","Stumpings")
from collections import namedtuple
IPL_Header = namedtuple('IPL_Header',fields)
def parseRecs(line):
    fields = line.split(",")
    return IPL_Header(fields[0],fields[1],fields[2],fields[3],fields[4],fields[5],fields[6],fields[7],fields[8],fields[9],fields[10],fields[11],fields[12],fields[13],fields[14],fields[15],fields[16],fields[17],fields[18],fields[19],fields[20],fields[21],fields[22],fields[23],fields[24],fields[25],fields[26],fields[27],fields[28])
```

```python
ipl_data = ipl_data_rdd.map(lambda line: parseRecs(line))
```

```python
ipl_data.take(2)
```

```python
#Count Number of mataches player by each plyers.
ipl_data.map(lambda rec: (rec.Player, rec.Matches)).reduceByKey(lambda a,b: int(a)+int(b)).collect()
```

```python
# No Of Matches Played by 'MS Dhoni'
ipl_data.filter(lambda rec: rec.Player=='MS Dhoni').map(lambda rec: (rec.Player,rec.Matches)).reduceByKey(lambda a,b: int(a)+int(b)).collect()
```

```python
total_matched_played_each_player = ipl_data.map(lambda rec: (rec.Player,rec.Matches)).reduceByKey(lambda a,b: int(a)+int(b))
# No Of Matches Played by 'MS Dhoni'
total_matched_played_each_player.filter(lambda rec: rec[0]== 'MS Dhoni').collect()
```

```python
# List of Player who player more than 50 Matches
total_matched_played_each_player.filter(lambda rec: int(rec[1]) > 50).collect()
```

```python
#Count Number of mataches player by each plyers and arrange in sorrted based on most no of maches played.
ipl_data.map(lambda rec: (rec.Player,rec.Matches)).reduceByKey(lambda a,b: int(a)+int(b)).sortBy(lambda r: (int(r[1])), False).collect()
```

```python
#Count the number of total centtury made by each player
ipl_data.map(lambda rec: (rec.Player,(0 if rec.Century=='-' else rec.Century))).reduceByKey(lambda a,b: int(a)+int(b)).sortBy(lambda r: (int(r[1])), False).collect()
```

```python
# Find the total number of century scored by 'Virat Kohli'
ipl_data.filter(lambda rec: rec.Player=='V Kohli').map(lambda rec: (rec.Player,(0 if rec.Century=='-' else rec.Century)) ).reduceByKey(lambda a,b: int(a)+int(b)).collect()
```

```python
# Find Player who scored most runs as total in IPL
ipl_data.map(lambda rec: (rec.Player,(0 if rec.Runds_Scored=='-' else rec.Runds_Scored))).reduceByKey(lambda a,b:(int(a)+int(b))).max(lambda r: int(r[1]))
```

```python
# Find player who scored highest in 1 inning.
def preprocess_highest_score(rec):
    if rec.Highest_Score.endswith("*"):
        return rec.Highest_Score.replace("*","")
    return 0 if rec.Highest_Score=='-' else rec.Highest_Score
    
ipl_data.map(lambda rec: (rec.Player, preprocess_highest_score(rec))).max(lambda r: int(r[1]))
```

```python
# Find Player in order to batting averge.
ipl_data.max(lambda rec: (0 if rec.Batting_Average=='-' else float(rec.Batting_Average)))
```

```python
# Find Player in order to batting Strike Rate is Heighest.
ipl_data.max(lambda rec: (0 if rec.Batting_Strike_Rate=='-' else float(rec.Batting_Strike_Rate)))
```
